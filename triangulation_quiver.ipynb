```python
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import matplotlib.tri as Tri
import pyugrid
import cartopy.crs as ccrs
import numpy as np
%matplotlib inline
```

```python
url = "http://comt.sura.org/thredds/dodsC/data/comt_1_archive/inundation_tropical/UND_ADCIRC/Hurricane_Ike_3D_final_run_with_waves"
```

```python
nc = Dataset(url,'r')
variables = ['u_vel','v_vel']
```

```python
nc.variables.keys()
```

```python
times = nc.variables['time'][:]
```

```python
times
```

```python
u = nc.variables['u-vel'][0]
v = nc.variables['v-vel'][0]
```

```python
print u.shape
print v.shape
```

```python
ug = pyugrid.UGrid.from_nc_dataset(nc)
lat = ug.nodes[:,0]
lon = ug.nodes[:,1]
```

```python
print lat.shape
print lon.shape
```

```python
plt.quiver(lon,lat, u, v)
```

```python
mag = np.sqrt(np.sum(np.square(np.asarray([u, v])),0))
```

```python
mag.shape
```

```python
plt.quiver(lon,lat,u,v,mag)
```

```python
plt.quiver(lat,lon,u,v,mag)
```
