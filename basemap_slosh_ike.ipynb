```python
import os
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.use("Agg")
from mpl_toolkits.basemap import Basemap
from matplotlib.figure import Figure
import matplotlib.pyplot as plt
from matplotlib.backends.backend_agg import FigureCanvasAgg
import matplotlib.tri as Tri
import numpy as np
import cartopy.crs as ccrs
from netCDF4 import Dataset

import pyugrid
%matplotlib inline

lonmin = -116.166686996
latmin = 9.93907265811
lonmax = -65.5856323151
latmax = 39.5485138771
```

```python
url = "http://comt.sura.org/thredds/dodsC/data/comt_1_archive/inundation_tropical/MDL_SLOSH/Hurricane_Ike_final_run_egm3"
```

```python
nc = Dataset(url,'r')
```

```python
lon = nc.variables['lon'][:]
lat = nc.variables['lat'][:]
data = nc.variables['etamax'][:]
```

```python
print "min(lon) = {0}, max(lon) = {1}".format(np.min(lon),np.max(lon))
print "min(lat) = {0}, max(lat) = {1}".format(np.min(lat),np.max(lat))
print "min(data) = {0}, max(data) = {1}".format(np.min(data), np.max(data))
print "lon.shape = {0}, lat.shape = {1}, data.shape = {2}".format(lon.shape,lat.shape,data.shape)
```

```python
plt.contourf(lon,lat,data)
```

```python
# create polar stereographic Basemap instance.
fig = plt.figure(figsize=(8,8))

m = Basemap(projection='merc',
            llcrnrlon=lonmin,
            llcrnrlat=latmin,
            urcrnrlon=lonmax,
            urcrnrlat=latmax,
            resolution='l')
m.ax=fig.add_axes([0,0,1,1])
cs = m.ax.contourf(lon,lat,data,15,cmap=plt.cm.jet)
m.drawcoastlines()

plt.show()
```
